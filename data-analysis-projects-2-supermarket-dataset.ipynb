# Data Visualization

```python
#Checking correlation between columns within the dataset
sns.heatmap(df.corr())
```

It can be seen from data description and the heatmap that gross margin percentage has same value througout the dataset, since, there is no relation in between. So, we are droping **'gross margin percentage'** column

```python
df.drop(columns=['gross margin percentage'], axis = 1, inplace=True)
```
